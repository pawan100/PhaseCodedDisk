```python
# import libraries
import numpy as np
import imageio
```

```python
# Create a 2D numpy array for the binary image
binary_image = np.zeros((30, 30), dtype=np.uint8)
```

```python
print(binary_image)
```

```python
# Set specific rows and columns to 1 (white) using loops
for col in range(30):
    for row in range(max(0, col), min(col + 3, 30)):
        binary_image[row, col] = 1
```

```python
print(binary_image)
```

```python
# Save the binary image as a TIF file
imageio.imwrite('binary_image1.tif', binary_image)
```

```python
for col in range(29, -1, -1):  # Columns 29 to 0 (in reverse order)
    for row in range(max(0, 29-col), min(30-col, 15), 2):
        binary_image[row, col] = 1
        binary_image[row+1, col] = 1
```

```python
print(binary_image)
```

```python
# Save the binary image as a TIF file
imageio.imwrite('binary_image2.tif', binary_image)
```

```python
# Set specific rows and columns to 1 (white) using loops
binary_image = np.zeros((30, 30), dtype=np.uint8)
for col in range(30):
    for row in range(max(0, col), min(col + 6, 30)):
        binary_image[row, col] = 1
```

```python
print(binary_image)
```

```python
# Save the binary image as a TIF file
imageio.imwrite('binary_image3.tif', binary_image)
```

```python
for col in range(29, -1, -1):  # Columns 29 to 0 (in reverse order)
    for row in range(max(0, 29-col), min(30-col, 15), 3):
        binary_image[row, col] = 1
        binary_image[row+1, col] = 1
        binary_image[row+2, col] = 1
```

```python
print(binary_image)
```

```python
# Save the binary image as a TIF file
imageio.imwrite('binary_image4.tif', binary_image)
```
